# 1. Introduction to Sustainable Development Goals (SDGs)


## 1.1 Sustainable Development and 2030 Agenda

Sustainable development can be defined as the development that meets the needs of the present without compromising the ability of future generations to meet their own needs.  
  
Sustainable development requires joint efforts across three core dimensions: economic growth, social inclusion and environmental protection. All these three dimensions are equally important, and sustainable development requires balanced approach across all three.
In 2015, the United Nations (UN) and its member states adopted 2030 Agenda and 17 Sustainable Development Goals to reaffirm their commitment to sustainable development and guide efforts towards universal peace and prosperity.  
  

## 1.2 What are SDGs?

The Sustainable Development Goals (SDGs) or the Global Goals, were adopted by the United Nations in 2015 as a universal call to action to end poverty, protect the planet, and ensure that by 2030 all people enjoy peace and prosperity.  
  
The Global Goals are an indispensable part of 2030 Agenda, providing a measurable and actionable plan. The 17 Global Goals are integrated and indivisible, and also balance economic, social, and environmental dimensions of sustainable development.

## 1.3 The importance of Sustainable Development and SDGs

The Development Goals are universal and apply to all countries regardless of their income and development level to address global challenges such as poverty, climate change and others. Moreover, the SDGs follow the guiding principle of “leaving no one behind”, which focuses on addressing the needs of women, children, and other vulnerable groups.  
  
The SDGs are not legally binding, but they guide national and global efforts toward sustainable development and achievement of the 17 Goals. There has been a strong trend both for national governments and corporate sector to report on the SDGs. National governments are encouraged to regularly review and report their progress through [Voluntary National Review mechanism](https://sustainabledevelopment.un.org/vnrs/). Similarly, enterprises are encouraged to reflect their contribution to the achievement of the SDGs in their Corporate Social Responsibility Reports, the [UN Global Compact and Global Reporting Initiative (GRI)](https://www.unglobalcompact.org/take-action/action-platforms/sdg-reporting) provides guidance and support on corporate SDGs reporting.


## 1.4 Challenges in achieving the SDGs

The achievement of the SDGs requires resources and innovative solutions from actors across all sectors as well as stronger collaboration between them.  
  
As sustainable development entails holistic and inclusive approach, it involves many different stakeholders, including national governments, the private sector, NGOs and others. Often, those stakeholders would have different interests and achievement of the SDGs requires constant negotiations between them as well as strong commitment for sustainable development from each stakeholder.  
  
Even though SDGs are indivisible, potential trade-offs exist between them. For instance, water security and biodiversity can be negatively affected with the expansion of agricultural lands to combat hunger and poverty. Thus, it is important to recognize those connections between the goals and address them accordingly and make sure that no one is left behind .   
  
Accountability and reporting the progress towards the achievement of the SDGs also possess a major challenge to the stakeholders. Monitoring the activities and progress at both national and global levels from different stakeholders is particularly difficult as the data is coming from different sources and in different formats.

# 2. The role of frontier technologies in achieving the SDGs

Frontier technologies such as Artificial Intelligence, Machine Learning, Big Data, GIS and other offer innovative solutions for major humanity challenges such as poverty, global warming, and others. Moreover, rapid technological advancement and lowering costs can help enable developing countries to “catch up”, and benefit from those technologies to accelerate the achievement of the SDGs.

## 2.1 What are frontier technologies?

There is no generally accepted definition of frontier technologies but for this course we will define them as emerging innovative technologies, which have the potential to significantly influence societies, economies, the environment and other areas. The scope of the frontier technologies is vast as they include Artificial Intelligence (AI), Machine Learning (ML), Geographic Information System (GIS), Blockchain, 3D printing, and many others. Those technologies evolve at an unprecedented rate, and it is becoming a challenge to track their development and application to the SDGs. Thus, this course will cover 2 key frontier technologies, which are mature and have wide application in development context - Artificial Intelligence and its subfields of Machine Learning, Deep Learning and Natural Language Processing (NLP) and GIS.

## 2.2 Overview of key technologies

### 2.2.1 Artificial Intelligence

Artificial Intelligence (AI) is an umbrella term for machines’ abilities to acquire and apply knowledge, and to carry out intelligent behavior such as reasoning, decision-making, learning and others, normally requiring human intelligence  . The term encompasses a range of technologies such as Machine Learning, Deep Learning and Natural Language Processing and others, with some of them discussed in detail in this course. Overall, AI can provide innovative solutions to almost any field through its automation, big data analysis and other capacities.

#### 2.2.1.1 Machine Learning

Machine Learning (ML) is the subfield of AI, which builds on the idea of machines learning by themselves using thousands of examples . The use of Machine Learning can be used in prediction of natural hazards such as floods, earthquakes and others. For instance, [UNDP in Bosnia and Herzegovina](https://undp.cz/portfolio-item/computer-vision-flood-forecasting-using-remotely-sensed-data/) uses ML models to analyze the data obtained from satellite images to predict floods and support local decision-makers to respond.


**Supervised, Unsupervised Learning and Reinforcement Learning**

Rapid advancement in artificial intelligence made the field too wide. To specialize in Artificial Intelligence (AI) and Machine Learning (ML), approaches such as supervised and unsupervised learning are formed by scientists. They differ in their applications to various problems and other nuances. However, the fundamental difference can be defined as that one uses labeled data to assist in predicting possible outcomes, whereas the other does not.

**Supervised Learning**

This approach relies on labeled datasets to guide algorithms in making predictions or classifications. That means some data is already matched with the correct answer. It kind of resembles learning in front of a teacher. These datasets “supervise” algorithms while classifying data or predicting possible outcomes correctly. Moreover, the algorithm learns from labeled training data over time and measures its accuracy.
In supervised learning, it is beneficial that the algorithm learns and adjusts itself for accuracy. Therefore, outcomes tend to be correct; however, it requires human intervention.


**Unsupervised Learning**

This approach uses unlabeled data while to analyze and cluster datasets. It means that the model does not need to be supervised. In this approach, models work by themselves to discover information or hidden patterns without any intervention by humans.
In unsupervised learning, it is beneficial that the algorithm can find inherited patterns or structures. Also, it may only require some human intervention for output validation. However, it might be time-consuming due to the calculation and analysis of all possibilities by the algorithm.

**Reinforcement Learning**

Reinforcement learning (RL) is a subfield of machine learning where an agent learns to make decisions by performing actions in an environment to maximize cumulative rewards. Unlike supervised and unsupervised learning, RL involves learning from the consequences of actions. The agent receives rewards or penalties based on its actions, allowing it to learn optimal strategies through trial and error.


#### 2.2.1.2 Deep Learning

Deep Learning is a subfield of ML, which provides even more advanced solutions to the problems through imitating the work of the human brain by analyzing information inputs through multiple analysis layers and providing the outputs. For instance, [UNDP Tanzania](https://www.undp.org/tanzania/blog/using-artificial-intelligence-tackle-infodemic-tanzania) has applied Deep Learning to analyze Twitter discussions on Covid-19 involving thousands of tweets and identify the scope of misinformation spread.

#### 2.2.1.3 Natural Language Processing

Natural Language Processing (NLP) is a subfield of AI, which is aimed to process and analyze human language inputs .  One of the most common applications of the NLP in development are chatbots, which can be used to provide information to the public as well as provide some informational support through mobile apps. The prominent example of chatbots application is the launch of [WhatsApp chatbot by the World Health Organization (WHO)](https://www.whatsapp.com/coronavirus/who) to respond to the most common questions on Covid-19 .

### 2.2.2 GIS

Geographic Information System (GIS) technologies are used to collect, store, analyze, manage, and present spatial or geographical data . GIS is commonly used in disaster management contexts, as the technology offers solutions across all stages of the disaster cycle. GIS can be used to develop interactive maps for preparedness, response, recovery and prevention efforts. For example, UNDP Pakistan has developed an interactive map to analyze the earthquake consequences to provide necessary assistance and resources to the affected people.

## 2.3 Digital divide and use of technologies in development context

Despite tremendous technological progress and wide application of frontier technologies across the world, there is still a significant digital divide between developed and developing countries. Often developed countries and large tech companies are at the technology frontier, while most of the developing countries are yet to benefit from existing technologies. However, the main issue is that the development and application of the technologies is resource- and skills-intensive process, which makes it difficult for developing countries, which often lack robust infrastructure and modern educational system, to catch up with the progress. At the same time, developing countries need those technologies the most to advance sustainable development and achieve the Global Goals. Frontier technologies present developing countries with opportunities to leapfrog traditional development stages, accelerating progress toward achieving the SDGs.

For example:

1) Mobile Banking: In many developing countries, access to traditional banking services is limited, especially in rural areas. Mobile banking platforms like M-Pesa in Kenya allow people to perform financial transactions using their mobile phones, eliminating reliance on physical bank branches. This technology has improved financial inclusion, allowing individuals to save money, receive payments, and access credit, while also supporting economic growth and poverty reduction (SDG 1: No Poverty).

2) Telemedicine: Telemedicine provides remote medical consultations and diagnostic services through digital platforms, which is particularly valuable in areas with a shortage of healthcare facilities and professionals. In countries like India, telemedicine services have expanded access to healthcare for rural populations, reducing the need for long-distance travel to urban hospitals and improving health outcomes (SDG 3: Good Health and Well-being).

3) E-Learning: With limited access to quality education, especially in remote areas, e-learning platforms and digital classrooms can bridge the educational gap. Initiatives like the One Laptop per Child program aim to provide children in developing countries with access to low-cost laptops and educational software, enhancing learning opportunities and educational outcomes (SDG 4: Quality Education).

4) Precision Agriculture: Advanced technologies such as drones, sensors, and data analytics can optimize agricultural practices, increase crop yields, and reduce resource use. For example, smallholder farmers in developing countries are starting to use precision agriculture tools to monitor soil health and weather patterns, leading to more efficient and sustainable farming practices (SDG 2: Zero Hunger).

5) Blockchain for Supply Chains: Blockchain technology can enhance transparency and efficiency in supply chains, ensuring fair trade and reducing fraud. For instance, the use of blockchain in agricultural supply chains helps farmers receive fair prices for their produce by providing a transparent record of transactions, which can improve their livelihoods and promote sustainable agriculture (SDG 12: Responsible Consumption and Production).

6) Natural Hazard 4Rs: GIS can significantly enhance recovery and relief efforts after an earthquake, particularly in developing countries, by increasing their capabilities to cope with natural hazards. In 2015, Nepal experienced a devastating earthquake that caused extensive damage and loss of life. GIS was pivotal in the response and recovery efforts, showcasing its potential to enhance disaster management capabilities. By integrating GIS into earthquake response and recovery efforts, Nepal was able to improve its disaster management capabilities significantly. This not only enhanced the immediate response to the 2015 earthquake but also contributed to long-term resilience against future natural hazards (SD11: Sustainable Cities and Communities).

# 3. Responsible use of technologies for Sustainable Development

To make use of the frontier technologies for sustainable development and SDGs, it is necessary to understand the ongoing debates related to ethics and data privacy in tech industry.

## 3.1 Ethical issues and considerations

As frontier technologies are getting used in almost every field now, there has been a major debate on ethical issues related to their use as well as their governance. Technologies such as AI and its subfields are prone to the biases of the humans who developed them.
One of the recent cases covers the automated high school exams grading to replace in person exams in the UK due to Covid pandemic. The algorithm used students grades and the school track records. The produced results were biased against the outstanding students from minority and low-income neighborhoods, as their schools has lower average grades than the ones in wealthier neighborhoods. Once the issue was identified, there was significant societal uproar, leading to the cancellation of the results.
Thus, it is important to ensure that AI and other tech projects are developed and implemented by diverse teams, who can recognize and minimize biases, prejudice and discrimination. Moreover, there has been an ongoing effort led by [UNESCO](https://en.unesco.org/artificial-intelligence/ethics) on national and international AI standards to minimize technology biases and possible misuse.

## 3.2 Data Privacy

Another major concern regarding the use of technologies is data privacy. Most of the frontier technologies are powered by the vast amounts of data collected from different sources, including individuals. The main issue is the lack of a common understanding of the privacy-related consequences of sharing personal data, as well as the difficulty in predicting how data can be interpreted and used with the rapid advancements in machine learning and related technologies.

In this context, the European Union (EU) has adopted General Data Protection Regulation (GDPR), which protects data privacy and security of its citizens and residents. The regulation addressed the issues of personal data collection, processing. There has been growing recognition of the GDPR as the gold standard for data privacy regulations, and more countries are expected to adopt similar policies.

## 3.3. Responsible Use

However, all those issues are raised not to discourage the use of the frontier technologies but to ensure their responsible use for sustainable development and SDGs. Since there is no central authority imposing specific limits on the use and governance of data, parties holding the data are enabled to set their own rules. To address these challenges, tech companies, policymakers, and societies must recognize ethical issues and data privacy concerns, engaging in public discussions and implementing national and international regulations and standards.
