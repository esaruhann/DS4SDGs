Exploratory Data Analysis (EDA) is the initial step in analyzing your dataset and a crucial part of preparing your data before applying a predictive model. These steps are prerequisites for modeling.

### 2.1 Introduction to Pandas DataFrame and Series



A **Series** object in the pandas library is defined in the [documentation](https://pandas.pydata.org/docs/reference/api/pandas.Series.html) as a "one-dimensional ndarray with axis labels (including time series)."
- A Series object can be created from an array-like object (e.g., Python list, NumPy array), an iterable, a dict or a scalar value.
- Math operators (+, -, /, *, **) can be used to combine two Series objects regardless of their size.

A **DataFrame** object, on the other hand, is [defined](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html) as a "two-dimensional, size-mutable, potentially heterogeneous tabular data."
- A DataFrame object can be created from an n-dimensional array (structured or homogeneous), an iterable or a dict.




**Create Pandas Series and DataFrame**

To create a Pandas Series or DataFrame, or perform any operation using the library, we need to import it. This is true for any other python library.

So, let's import Pandas and demonstrate how to create a simple Pandas Series and DataFrame.

**Using `loc[]` function of the Dataframe**

The difference between `loc[]` vs `iloc[]` is how you select rows and columns from a pandas DataFrame:

* `loc[]` is used to select rows and columns by Names/Labels
* `iloc[]` is used to select rows and columns by Integer Index/Position. zero based index position.

One of the main advantages of pandas DataFrame is its ease of use. You can see this yourself when you use the `loc[]` or `iloc[]`methods to select or filter rows or columns. These are some of the most commonly used attributes in pandas DataFrame.

````python
print("Given Dataframe :\n", df)

print("\nIterating over rows using loc function :\n")

# iterate through each row and select
# 'area' and 'year' column respectively.
for i in range(len(df)):
    print(df.loc[i, "area"], df.loc[i, "year"])

    if i%10 == 0:
      break
````

**Using iterrows() method of the Dataframe**

Pandas `DataFrame.iterrows()` is used to iterate over DataFrame rows. This returns (index, Series) where the index is an index of the Row and Series is data or content of each row.

````python
print("\nIterating over rows using iterrows() method :\n")

counter = 0

# iterate through each row and select
# 'area' and 'year' column respectively.
for index, row in df.iterrows():

  #print only first 10 rows
    if counter < 10:
        print(row["area"], row["year"])
        counter += 1
    else:
        break

````

**Using itertuples() method of the Dataframe**

The `itertuples()` method of a pandas DataFrame is one of the most commonly used methods to iterate over rows, as it returns all DataFrame elements as an iterator containing a tuple for each row.

The main difference between this method and `iterrows()` is that `itertuples()` is faster. Additionally, it preserves the data type of each column, unlike iterrows(), which returns a Series for each row and does not preserve column data types. The itertuples() method returns a tuple where the first value is the index number of the row, and the remaining values are the values for each column

````python
print("\nIterating over rows using itertuples() method :\n")

counter = 0
# iterate through each row and select
# 'area' and 'year' column respectively.
for row in df.itertuples(index=True, name='Pandas'):
    if counter < 10:
        print(getattr(row, "area"), getattr(row, "year"))
        counter += 1
    else:
        break
````

## 4.1. Handling Missing Values

Missing data is actually the most common occurrence of dirty data. Imagine trying to do a customer segmentation analysis, but 50% of the data have no address on record. It would be hard or impossible to do your analysis since the analysis would be bias in showing no customers in certain areas. That is the reason why checking the missing data in our dataset is quite important.

Blank space, None, NaN, and NaT are the general types of missing data. Learning how to check the missing data is provided below.

### 4.1.1 df.isna() and df.isnull(): Check for Missing Values.



**`df.isna()`** returns a DataFrame of the same shape as the original, but with True in the location of elements that are NaN or None, and False elsewhere.

**`df.isnull()`** returns a DataFrame where each cell is either True or False depending on whether that cell's original value is missing (i.e., NaN in numerical arrays, None or NaN in object arrays, NaT in datetimelike). So if a cell's value is missing, df.isnull() will return True for that cell; otherwise, it will return False.

````python
# our sample data set has no missing values
df.isna()
````

````python
# check if dataset has null values
print(df.isnull())
````

````python
# print the number of missing values per column
print(df.isnull().sum())
````

### 4.3.9. df.merge():  

Allows you to join two dataframes together based on common columns

The type of merge to be performed is specified using the how parameter.

Here are the basic types of merges:

* left: Use keys from the left frame only.
* right: Use keys from the right frame only.
* inner: Use intersection of keys from both frames.
* outer: Use union of keys from both frames.

Let's illustrate each merging method using dataframes with columns "area", "year", and "life_expectancy".

````python
import pandas as pd

# Create two sample dataframes
df1 = pd.DataFrame({
    'area': ['Area1', 'Area2', 'Area3', 'Area4'],
    'year': [2001, 2002, 2003, 2004],
    'life_expectancy': [75, 78, 82, 85]
})

df2 = pd.DataFrame({
    'area': ['Area3', 'Area4', 'Area5', 'Area6'],
    'year': [2003, 2004, 2005, 2006],
    'life_expectancy': [83, 86, 79, 80]
})

print("DataFrame 1:")
print(df1)

print("\nDataFrame 2:")
print(df2)
````

````python
# Left Merge

left_merge = df1.merge(df2, on=['area', 'year'], how='left', suffixes=('_df1', '_df2'))
print(left_merge)
````

````python
# Right Merge

right_merge = df1.merge(df2, on=['area', 'year'], how='right', suffixes=('_df1', '_df2'))
print(right_merge)
````

````python
# Inner Merge

inner_merge = df1.merge(df2, on=['area', 'year'], how='inner', suffixes=('_df1', '_df2'))
print(inner_merge)
````

````python
# Outer Merge:
outer_merge = df1.merge(df2, on=['area', 'year'], how='outer', suffixes=('_df1', '_df2'))
print(outer_merge)
````

## 5. Outlier Detection and Treatment

An outlier is data that **differ significantly** from the other observations. As a result, distribution is found to be highly skewed because of the outliers, which can have harmful consequences on later use of the data, in Machine Learning, for example.
<br> Noticing and cleaning them is therefore an important part of the data exploration process.

**What cause an outlier?**

Outliers can indicate 2 things:
- The outliers can be due to a measurement error, such as a typo or a malfunction in the measurement tool. These should be cautiously discarded from the dataset!
- The outliers can result from a highly skewed (asymmetric) distribution or a heavy-tailed distribution. These situations can occur in non-normal distributions, such as exponential, geometric, Poisson, or bimodal distributions

*Assuming we are in normal distribution we mainly focus on the first case scenario and the dataset need to be cleaned up*

**How to detect an outlier?**
- Thanks to data visualization, we can easily identify outlier candidate as they stand far from the rest of the data point. However, this is not sufficient to just eyeball them, we need to test if an observation is indeed an outlier or not.

<br>There exist multiple mathematical methods to identify outliers:
- A data point can be defined as an outlier if its **distance to the mean is superior to 3 times the standard deviation**. Or in other term, when a data point **Z score** is > 3, it is considered an outlier.
Where:

<br><center>$Z score = (X - {\mu}) / {\sigma}$
- Similarly to the previous method, you may have heard of the 1.5 IQR method, which stands for 1.5 times the interquartile range (the range between the first quartile and the third quartile). This methodology yields similar results to the previous method, as 1.5 IQR is roughly equivalent to 3 standard deviations.

<br><center> Lower bound: $ Q1 - 1.5 * IQR$
<center> Upper bound: $ Q3 - 1.5 * IQR$

<br> - There exist more complex test that can attest outliers, like for example: the **[Local outlier method](https://scikit-learn.org/stable/auto_examples/neighbors/plot_lof_outlier_detection.html#:~:text=The%20Local%20Outlier%20Factor%20(LOF,lower%20density%20than%20their%20neighbors.)** or the **[Minimum covariance determinant](https://scikit-learn.org/stable/modules/generated/sklearn.covariance.MinCovDet.html)**.

### 6.1.1 Histogram Charts

Histograms are a graphical display of frequencies. They are represented as bars. They show what portion of the dataset falls into each category, usually specified as non-overlapping intervals. These categories are called bins. The .hist() method would help us to create a simple histogram.

To begin plotting our histogram using Matplotlib, we first of all need to import the package as indicated in the cell below. Then, call the hist() method on our data



````python
# Import Matplotlib
import matplotlib.pyplot as plt

data1 = np.random.randn(1000) #A sample generation based on random

plt.hist(data1)
plt.xlabel("x")
plt.ylabel("Frequency")
plt.show()
````

### 6.1.2 Bar charts

Bar charts display rectangular bars either in vertical or horizontal form. Their length is proportional to the values they represent. They are used to compare two or more values. The `.bar()` method would be used to create a bar graph.

### 6.1.3. Scatter Plot

Another commonly used plot type is the scatter plot. Here the points are represented individually with a dot or a circle. The `.scatter()` method would help to create a scatter plot.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2, linestyle='-', marker='o', markersize=6, markerfacecolor='red', markeredgewidth=2, markeredgecolor='black')

# Title
plt.title('Sine Function')  # Sets the title of the plot

plt.xlabel('x')
plt.ylabel('sin(x)')

# Customizing Tick Marks
plt.xticks(np.arange(0, 11, step=1))  # Set tick marks on the x-axis from 0 to 10
plt.yticks(np.arange(-1, 1.5, step=0.5))  # Set tick marks on the y-axis from -1 to 1.5

# Grid
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines with dashed lines and transparency

# Legend
plt.legend(loc='upper right')  # Add a legend at the upper right corner

plt.savefig('sin_plot.png')
plt.show()
````

### 6.1.4. Pie Charts

Pie charts are circular representations divided into sectors. The sectors are also called wedges. The `pie()` function is used to plot pie charts from an array X. Wedges are created proportionally, so that each value x of array X generates a wedge proportional to x/sum(X).


````python
labels = 'Health', 'Climate Change', 'Obesity'
sizes = [70.4, 20.6, 9.0]
colors = ['lightgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=True, startangle=90)
plt.title("What is the most concerning topic for the next 10 years?(100 people)")
plt.legend(title="Answers")
plt.axis('equal')
plt.show()
````

### 6.4.1 Kernel Density Plot  <a class="anchor" id="four-four-one"></a>

Kernel Density Estimation (KDE) is a method used to estimate the probability density function of a numerical variable, using a kernel function. We encourage the reader to dive deeper into the mathematical definitions and examples provided [here](https://machinelearningmastery.com/probability-density-estimation/), [here](https://www.statsmodels.org/stable/examples/notebooks/generated/kernel_density.html) and the [Wikipedia Article](https://en.wikipedia.org/wiki/Kernel_density_estimation). However, for now, it suffices that one can learn to read and interpret KDE plots, as the mathematical ground is well-established and several python libraries provide the output required.


KDE plots use a kernel, a mathematical function, to estimate the probability that a given value appears in the random variable being analyzed. Depending on the data being analyzed, different kernels and smoothing parameters should be chosen. Similar to histograms, the quality of the representation also depends on selecting appropriate smoothing parameters. For further guidance on KDE parameters and how to properly represent the data, refer to the Seaborn documentation. [here](https://seaborn.pydata.org/generated/seaborn.kdeplot.html).

As summary, KDE plots can be helpful to:
* See variable value range, i.e. min, max, mean
* See most common values in our data (peaks)
* Understand if our data is representative of all use cases we might be interested in, i.e. one may notice, by the peaks and valleys, that not enough samples of a particular value range were collected
* For classification problems, it is interesting to see how the variable curve shape might change according to the class
* Variables that do not present a Gaussian-like distribution (i.e. Heavy-Tailed) might need to through some kind of transformation (log or power transform) to improve model performance

## 7. Correlation and Covariance Analysis

Correlation and covariance are powerful statistical concepts used to measure the relationship between two or more variables.

### 7.1 Correlation Analysis and Visualization  <a class="anchor" id="four-four-one"></a>

A correlation coefficient (commonly represented by *r*) is a value that describes the strength of the linear relationship between two variables. A common method to measure correlation is the [Pearson correlation coefficient](https://www.statisticshowto.com/probability-and-statistics/correlation-coefficient-formula/#Pearson). The output is a value that can range from +1 to -1, representing the correlation between two features of our data. If two variables have a high correlation value, it means that both variables are "moving" in the same direction, i.e., they change together at a constant rate. For example, if one variable increases its value, the other also increases. Negative values represent variables moving in the opposite direction, i.e., they are inversely correlated. A value of r = 0 indicates no correlation (the variables are independent).

**What can we learn from the Correlation Heatmap?**
* For regression problems, variables with a high correlation (positive or negative) to the target value are usually important features for your dataset. In short, this means that when this feature increases, your target value will likely exhibit the same behavior.

* Correlation analysis can help identify non-relevant features or reduce the number of variables. For example, when two variables are highly correlated, they might be providing similar information to your model. In such cases, one of them can be removed to reduce the complexity of your data.

**What we should not assume from the Correlation Heatmap?**

* Variables that have values close to +1 and -1 present a **linear** relationship. When the correlation coefficient is close to 0, it does not mean that there is not some type of relationship there. It simply means that the relationship might be of another form (exponential, quadratic or something else). Use scatter plots to visualise the relationship between variables as well

* Correlation does not imply causation. From the correlation analysis, we cannot assume cause-and-effect between two variables. Using an example from the [Wikipedia](https://en.wikipedia.org/wiki/Correlation_does_not_imply_causation) article, let's say you analysing a study that highly correlates time watching TV and the level of aggressiveness in children. One could assume "Children that watch a lot of TV are the most violent. TV makes children more violent." However, this could be the other way round; that is, violent children like watching more TV than less violent ones.

Correlation is usually plotted in the form of a heatmap. The heatmap will have the maximum values on its diagonal, as it is the relationship between the variable and itself. As such, for the analysis, we only consider half the heatmap. Below we plot the correlation output of the iris dataset, using the command **.corr()** and the Seaborn library.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset as a DataFrame
iris_data = load_iris(as_frame=True)
iris = iris_data.frame

# Drop the non-numeric column ('target' is the species column in sklearn's iris dataset)
numeric_iris = iris.drop(columns=["target"])

# Calculate the correlation matrix
corr = numeric_iris.corr()

# Plot the heatmap
ax = sns.heatmap(corr, linewidths=.5, annot=True, fmt=".2f", cbar=False)
plt.show()

````

From the above plot, we can extract the following insights:
* Sepal Length seems to increase with Petal Length (0.87) and Petal Width (0.82). However, the Sepal Width is not strongly correlated (-0.12) to the Sepal Length
* Sepal Width is inversely correlated to Petal Length and Petal Width. One can think of this as: the larger the flower Sepal Width, it might present smaller Petal Lengths and Widths
* Petal Length and Width are highly correlated, close to 1. As such, increase in Petal Length is usually also reflected in an increase in Petal Width

For brevity, not all visualisations tools were demonstrated here (e.g. violin and box-plots). However, it gives the initial tools to start your data analysis and exploration journey.

## 7.2 Covariance

Covariance is a measure of how much two random variables vary together. It's used to gauge the linear degree of change between two variables, to see how much one can be predicted from the other. The sign of the covariance shows the tendency in the linear relationship between the variables:

Positive covariance: Indicates that two variables tend to move in the same direction.
Negative covariance: Reveals that two variables tend to move in inverse directions.

The example code cell shows how to compute covariance with the iris dataset and visualization using seaborn heatmap

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset from seaborn
iris = sns.load_dataset("iris")

# Exclude the non-numeric 'species' column
numeric_iris = iris.drop(columns=["species"])

# Compute the covariance matrix
covariance_matrix = numeric_iris.cov()

# Display the covariance matrix
print(covariance_matrix)

# Create a heatmap from the covariance matrix
plt.figure(figsize=(10, 8))
sns.heatmap(covariance_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Covariance Matrix Heatmap")
plt.show()
````
